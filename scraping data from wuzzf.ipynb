```python
for container in containers : 
    
    job_title = container.findAll("h2" , {"class" : "css-m604qf"})
    jtitle = job_title[0].text.strip()
    
    Company_name = container.findAll("a" , {"class" : "css-17s97q8"})
    cname = Company_name[0].text.strip()
    
    job_type = container.findAll("span" , {"class" : "css-1ve4b75 eoyjyou0"})
    jtype = job_type[0].text.strip()
    
    location = container.findAll("span" , {"class" : "css-5wys0k"})
    locjob = location[0].text.strip()
    
    level_of_experience = container.find_all("a" , {"class" : "css-o171kl"})
    levelexp = level_of_experience[1].text.strip()
    
    
    f.write(jtitle + "," + cname + "," + jtype + "," + locjob + "," + levelexp + "\n")
f.close()
```
